```python
import pandas as pd
import numpy as np
```

```python
data ={
    "Student_ID":[1,2,3,4,5,6,7,8],
    "Math_Score":[78,85,np.nan,90,45,10,67,88],
    "Science_Score":[72,80,75,np.nan,45,85,55,65],
    "English_Score":[65,70,68,75,np.nan,85,60,90],
    "Attendance_Percentage":[85,90,88,92,70,np.nan,80,99]
}
df=pd.DataFrame(data)
df
```

```python
df.fillna(df.mean(),inplace=True)
```

```python
df=df.clip(0,100)
```

```python
def iqr_fix(col):
  Q1 = col.quantile(0.25)
  Q3 = col.quantile(0.75)
  IQR =Q3 - Q1
  return col.clip(Q1-1.5*IQR,Q3+1.5*IQR)

  for col in ["Math_Score","Science_Score","English_Score"]:
    df[col]=iqr_fix(df[col])
```

```python
df["Attendance_Percentage"]=np.log1p(df["Attendance_Percentage"])
```
